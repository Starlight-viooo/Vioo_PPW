## Preprocessing

```python



import pandas as pd
import nltk
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Unduh sumber daya NLTK yang diperlukan (cukup sekali)
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('punkt_tab')

print("✅ Library dan sumber daya NLTK siap digunakan!")
```

```python
# Ganti dengan path file Anda jika perlu
NAMA_FILE_INPUT = 'berita_tempo_new.csv'

try:
    df = pd.read_csv(NAMA_FILE_INPUT)
    print(f"Berhasil memuat {len(df)} baris data dari '{NAMA_FILE_INPUT}'.")

    # Tampilkan 5 baris pertama untuk melihat struktur datanya
    print("\n--- 5 Baris Pertama Data Asli ---")
    display(df.head())

except FileNotFoundError:
    print(f"❌ ERROR: File '{NAMA_FILE_INPUT}' tidak ditemukan. Pastikan file berada di folder yang sama.")
```

```python
import pandas as pd
```

```python
def clean_text(text):
    # 1. Mengubah teks menjadi huruf kecil (lowercase)
    text = text.lower()
    # 2. Menghilangkan angka
    text = re.sub(r"\d+", "", text)
    # 3. Menghilangkan tanda baca (punctuation)
    text = text.translate(str.maketrans("","",string.punctuation))
    # 4. Menghilangkan spasi putih berlebih
    text = text.strip()
    return text

print("Menerapkan pembersihan teks...")
# Buat kolom baru 'isi_bersih' untuk menampung hasil pembersihan
df['isi_bersih'] = df['isi'].apply(clean_text)

print("✅ Pembersihan teks selesai.")

# Tampilkan perbandingan sebelum dan sesudah dibersihkan
print("\n--- CONTOH PERBANDINGAN ---")
print("Teks Asli:\n", df['isi'].iloc[0][:200], "...")
print("\nTeks Bersih:\n", df['isi_bersih'].iloc[0][:200], "...")
```

```python
# --- Tahap 2: Tokenisasi (Tokenization) ---

# Unduh paket 'punkt' yang diperlukan oleh NLTK untuk tokenisasi
# Langkah ini akan memastikan model tokenizer tersedia
try:
    nltk.data.find('tokenizers/punkt')
except nltk.downloader.DownloadError:
    print("Paket 'punkt' belum terunduh. Mengunduh sekarang...")
    nltk.download('punkt')
    print("✅ Paket 'punkt' berhasil diunduh.")

print("\nMenerapkan tokenisasi...")

# Terapkan tokenisasi pada kolom 'isi_bersih'
df['tokens'] = df['isi_bersih'].apply(word_tokenize)

print("✅ Tokenisasi selesai.")

# Tampilkan hasil tokenisasi untuk baris pertama
print("\n--- CONTOH HASIL TOKENISASI ---")
print(df['tokens'].iloc[0])
```

```python
# Muat daftar stopword Bahasa Indonesia dari NLTK
list_stopwords = stopwords.words('indonesian')

def remove_stopwords(tokens):
    return [word for word in tokens if word not in list_stopwords]

print("Menghapus stopwords...")
# Terapkan fungsi penghapusan stopwords
df['tokens_tanpa_stopword'] = df['tokens'].apply(remove_stopwords)

print("✅ Penghapusan stopwords selesai.")

# Tampilkan hasil setelah stopwords dibuang
print("\n--- CONTOH SETELAH STOPWORDS DIHAPUS ---")
print(df['tokens_tanpa_stopword'].iloc[0])
```

```python
# Inisialisasi stemmer
stemmer = PorterStemmer()

def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]

print("Menerapkan stemming...")
# Terapkan fungsi stemming
df['tokens_stemmed'] = df['tokens_tanpa_stopword'].apply(stem_tokens)

print("✅ Stemming selesai.")

# Tampilkan hasil setelah stemming
print("\n--- CONTOH SETELAH STEMMING ---")
print(df['tokens_stemmed'].iloc[0])
```

```python
# --- Tahap 5: Mengambil Kata Unik (Vocabulary) ---


df['kata_unik'] = df['tokens_stemmed'].apply(lambda tokens: sorted(list(set(tokens))))

print("✅ Proses pengambilan kata unik selesai.")

# Tampilkan perbandingan antara token sebelum dan sesudah dijadikan unik
print("\n--- CONTOH TOKEN SEBELUM DIJADIKAN UNIK ---")
print(df['tokens_stemmed'].iloc[0])

print("\n--- CONTOH KATA UNIK (VOCABULARY) ---")
print(df['kata_unik'].iloc[0])
```

```python
# --- Menyimpan Hasil Akhir (Kata Unik) ---

#    Ini agar mudah disimpan dan dibaca di dalam file CSV.
df['kata_unik_final'] = df['kata_unik'].apply(lambda tokens: ' '.join(tokens))

# 2. Pilih kolom-kolom relevan yang ingin disimpan
df_final = df[['id', 'judul_berita', 'kategori', 'kata_unik_final']]

# 3. Tentukan nama file output yang baru
NAMA_FILE_OUTPUT = 'berita_tempo_unik.csv'

# 4. Simpan DataFrame ke file CSV baru
df_final.to_csv(NAMA_FILE_OUTPUT, index=False, encoding='utf-8-sig')

print(f"✨ Proses selesai! Daftar kata unik telah disimpan di '{NAMA_FILE_OUTPUT}'")

# Tampilkan 5 baris pertama dari file hasil akhir untuk verifikasi
print("\n--- Tampilan 5 Baris Pertama dari File Kata Unik ---")
display(df_final.head())
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# --- 1. Memuat Data yang Sudah Diproses ---
NAMA_FILE_PROCESSED = 'berita_tempo_unik.csv'

try:
    df = pd.read_csv(NAMA_FILE_PROCESSED)
    # Ambil kolom 'isi_final' yang berisi teks bersih dan gabungkan menjadi sebuah list
    # Kita juga perlu menangani baris yang mungkin kosong (NaN) dengan mengubahnya menjadi string kosong
    corpus = df['kata_unik_final'].fillna('').tolist()
    print(f"✅ Berhasil memuat {len(corpus)} dokumen dari '{NAMA_FILE_PROCESSED}' untuk perhitungan TF-IDF.")

except FileNotFoundError:
    print(f"❌ ERROR: File '{NAMA_FILE_PROCESSED}' tidak ditemukan. Pastikan Anda sudah menjalankan sel preprocessing sebelumnya.")
    corpus = [] # Buat list kosong agar tidak error

# --- 2. Menjalankan TF-IDF ---
if corpus:
    # Inisialisasi TfidfVectorizer
    vectorizer = TfidfVectorizer()

    # Lakukan fit_transform untuk mempelajari vocabulary dan menghitung bobot TF-IDF
    X = vectorizer.fit_transform(corpus)

    # --- 3. Menampilkan Hasil ---
    print("\n--- Hasil Perhitungan TF-IDF ---")

    # Tampilkan daftar kata (fitur) yang berhasil diidentifikasi
    print("Contoh beberapa fitur (kata) yang ditemukan:")
    print(vectorizer.get_feature_names_out()[:20]) # Tampilkan 20 kata pertama

    # Tampilkan dimensi dari matriks TF-IDF yang dihasilkan
    print("\nDimensi Matriks TF-IDF (dokumen, fitur):")
    print(X.shape)

    # Menampilkan matriks dalam bentuk DataFrame agar mudah dibaca (opsional, bisa sangat besar)
    print("\n--- Contoh Matriks TF-IDF (5 dokumen pertama, 10 fitur pertama) ---")
    df_tfidf = pd.DataFrame(X.toarray()[:5, :10], columns=vectorizer.get_feature_names_out()[:10])
    display(df_tfidf)
```

```python
from gensim.models import Word2Vec
import numpy as np
import pandas as pd

# --- 1. Memuat Data Unik ---
NAMA_FILE_UNIK = 'berita_tempo_unik.csv'

try:
    df_processed = pd.read_csv(NAMA_FILE_UNIK)
    print(f"✅ Berhasil memuat data dari '{NAMA_FILE_UNIK}'")
except FileNotFoundError:
    print(f"❌ ERROR: File '{NAMA_FILE_UNIK}' tidak ditemukan. Pastikan Anda sudah menjalankan sel preprocessing sebelumnya.")
    raise # Hentikan eksekusi jika file tidak ada

# --- 2. Mempersiapkan Data untuk Model ---

# Nama kolom yang berisi teks bersih (kata-kata unik yang digabung)
kolom_teks = 'kata_unik_final'

if kolom_teks not in df_processed.columns:
    print(f"❌ ERROR: Kolom '{kolom_teks}' tidak ditemukan dalam dataframe.")
else:
    # Word2Vec membutuhkan input berupa list dari list token.
    # Kita ubah setiap kalimat di kolom 'kata_unik_final' menjadi list token dengan .split()
    sentences = df_processed[kolom_teks].fillna('').apply(lambda x: x.split()).tolist()

    print("\n=== Proses Pelatihan Word Embedding (Word2Vec) ===")

    # --- 3. Melatih Model Word2Vec ---
    # vector_size: dimensi vektor untuk setiap kata
    # window: ukuran jendela konteks (kata-kata tetangga)
    # min_count: abaikan kata yang muncul kurang dari nilai ini
    word2vec_model = Word2Vec(sentences=sentences, vector_size=100, window=5, min_count=1, workers=4)
    print("✅ Model Word2Vec berhasil dilatih.")

    # --- 4. Membuat Vektor Dokumen ---

    # Fungsi untuk mendapatkan vektor dokumen (rata-rata vektor kata)
    def get_document_vector(tokens, model):
        # Ambil vektor untuk setiap token yang ada di dalam vocabulary model
        vectors = [model.wv[token] for token in tokens if token in model.wv]
        if vectors:
            # Rata-ratakan semua vektor kata untuk mendapatkan satu vektor dokumen
            return np.mean(vectors, axis=0)
        else:
            # Kembalikan vektor nol jika tidak ada kata yang dikenali model
            return np.zeros(model.wv.vector_size)

    # Terapkan fungsi ke 'sentences' untuk mendapatkan vektor setiap berita
    df_processed['document_vector'] = [get_document_vector(doc, word2vec_model) for doc in sentences]

    # --- 5. Menampilkan Hasil ---

    # Buat DataFrame dari vektor dokumen agar mudah dibaca
    document_vector_df = pd.DataFrame(df_processed['document_vector'].tolist(),
                                      columns=[f'dim_{i}' for i in range(word2vec_model.wv.vector_size)])

    print("\n--- Contoh Vektor Dokumen (5 Berita Pertama) ---")
    display(document_vector_df.head())

    print("\n✨ Proses Word Embedding selesai.")
```
